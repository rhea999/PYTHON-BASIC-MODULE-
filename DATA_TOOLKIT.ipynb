# 1. What is NumPy, and why is it widely used in Python?

2.  How does broadcasting work in NumPy?

Broadcasting is a powerful feature in NumPy that allows arrays of different shapes to be used in arithmetic operations without explicitly reshaping or duplicating data.

Instead of looping through elements manually, NumPy automatically "broadcasts" the smaller array across the larger one so that they have compatible shapes.

Basic Broadcasting Rules
Two arrays are compatible for broadcasting if:

* They have the same shape, or

* The dimensions are equal to 1, or

* One of them is missing (i.e., smaller array)

Rule: Start comparing shapes from the trailing dimensions and work backward.

Example 1: Adding a scalar to an array

    import numpy as np

    a = np.array([1, 2, 3])
    b = 5

    # Broadcasts 5 to [5, 5, 5]
    result = a + b  # [6, 7, 8]

Example 2: 2D and 1D array addition

      a = np.array([[1, 2, 3],
                    [4, 5, 6]])
      b = np.array([10, 20, 30])

      # b is broadcast across each row of a
      result = a + b
      # [[11, 22, 33],
      #  [14, 25, 36]]

4.  Explain the use of the groupby() method in Pandas.

The groupby() method in Pandas is used to split data into groups based on some column(s), apply a function (like sum, mean, count, etc.), and then combine the results.

This is also known as the “Split–Apply–Combine” strategy.

Why Use groupby()?
* To analyze patterns within groups.

* To perform aggregations like total sales per region, average marks per student, etc.

* To summarize or segment data easily.

example:

        import pandas as pd

6. What are the differences between NumPy arrays and Python lists?

1. Data Type
* Python List: Can store mixed data types (e.g., integers, strings, floats).

* NumPy Array: Stores data of only one type (all integers, all floats, etc.).
2. Performance
* Python List: Slower for numerical operations.

* NumPy Array: Faster due to optimized C-based backend.
3. Memory Usage
* Python List: Uses more memory (stores object references).

* NumPy Array: Memory-efficient, stores elements in contiguous memory.
4. Functionality
* Python List: Limited built-in functions for math.

* NumPy Array: Has many mathematical functions (like sum(), mean(), sqrt()).

5. Mathematical Operations
* Python List: Operations like + concatenate lists.

* NumPy Array: Operations are element-wise (e.g., [1,2] + [3,4] = [4,6]).

6. Multidimensional Support
* Python List: Needs nested lists manually for 2D or 3D.

* NumPy Array: Built-in support for multidimensional arrays.

7. Broadcasting
* Python List: Does not support broadcasting.

* NumPy Array: Supports broadcasting (auto-expands shapes during operations).
 8. Use Case
* Python List: Best for general-purpose programming.

* NumPy Array: Ideal for scientific computing, data analysis, and machine learning.

15. How does NumPy handle multidimensional arrays?

NumPy handles multidimensional arrays using its core data structure called the ndarray (n-dimensional array). This allows efficient storage, manipulation, and computation on data of any number of dimensions — 1D, 2D, 3D, or more.

1. Creating Multidimensional Arrays

        import numpy as np

        # 2D array
        a = np.array([[1, 2], [3, 4]])

        # 3D array
        b = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])

2. Accessing Elements

        print(a[1][0])      # Output: 3 (from 2D)
        print(b[1][0][1])   # Output: 6 (from 3D)


3. Array Shape and Dimensions
* ndarray.shape: Returns the dimensions (rows, columns, depth, etc.)

* ndarray.ndim: Returns the number of dimensions

        print(a.shape)  # (2, 2)
        print(b.shape)  # (2, 2, 2)
        print(a.ndim)   # 2
        print(b.ndim)   # 3

4. Slicing Multidimensional Arrays
Just like lists, NumPy arrays can be sliced easily:

        print(a[:, 1])  # Second column of 2D array: [2 4]
        print(b[0, :, :])  # First "layer" of 3D array

5. Operations Are Element-Wise
You can apply operations across entire arrays without loops:

        c = a * 2
        print(c)  # [[2, 4], [6, 8]]

17. Explain the difference between apply() and map() in Pandas.


1. map()
* Used only on Series (1D).

* Applies a function, dictionary, or Series to each element in a Series.

* Element-wise transformation.

Example:

      import pandas as pd

      s = pd.Series([1, 2, 3])

      # Double each value
      s.map(lambda x: x * 2)
      # Output: 0    2
      #         1    4
      #         2    6

 2. apply()
* Can be used on both Series and DataFrame.

* On Series: behaves like map() (element-wise).

* On DataFrame: applies a function along rows or columns (axis=0 for columns, axis=1 for rows).

Example:-

        df = pd.DataFrame({
            'A': [1, 2],
            'B': [3, 4]
        })

        # Sum each row
        df.apply(sum, axis=1)
        # Output:
        # 0    4
        # 1    6

21. A Why is NumPy’s array slicing faster than Python’s list slicing?

NumPy's array slicing is significantly faster than Python's list slicing due to the following reasons:

1. Contiguous Memory Layout
* NumPy arrays store data in contiguous blocks of memory, just like C arrays.

* Python lists store pointers to objects, which are scattered in memory.

This makes NumPy operations much faster because memory access is efficient.

2. Fixed Data Types
* All elements in a NumPy array are of the same data type (homogeneous).

* Python lists are heterogeneous (can contain any type).

 NumPy doesn’t need to check the type of each element, which saves time.

3. View Instead of Copy
* Slicing a NumPy array returns a view (reference) to the original data.

* Slicing a Python list returns a new list (i.e., it copies the data).

 NumPy avoids unnecessary memory allocation and copying.

 Example:-

      import numpy as np

      a = np.array([1, 2, 3, 4])
      b = a[1:3]  # This is a view, not a copy
      b[0] = 99
      print(a)  # Output: [1 99 3 4]

4. Low-Level Optimizations
* NumPy is written in C and Fortran, which are compiled and optimized for performance.

* Python list slicing relies on higher-level Python code.

NumPy benefits from hardware-level speed.

```python
4. #How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame
data = {
    'Math': [80, 85, 78, 92, 88],
    'Science': [75, 90, 85, 95, 89],
    'English': [78, 82, 20, 85, 87]
}

df = pd.DataFrame(data)

# Calculate the correlation matrix
corr_matrix = df.corr()

# Create a heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)

# Add title
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
8. # Create a line plot with multiple lines using Matplotlib?

import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]

# Two different sets of y-values
y1 = [2, 3, 5, 8, 11]
y2 = [1, 4, 6, 8, 10]

# Plot the first line
plt.plot(x, y1, label='Line 1', color='blue', marker='o')

# Plot the second line
plt.plot(x, y2, label='Line 2', color='red', marker='x')

# Add title and labels
plt.title('Line Plot with Multiple Lines')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Add legend to distinguish lines
plt.legend()

# Show plot
plt.show()
```

```python
9. # Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold?

import pandas as pd

# Create a sample DataFrame
data = {
    'Name': ['Rhea', 'And', 'Old', 'beaue'],
    'Age': [25, 55, 29, 26],
    'Score': [88, 92, 79, 85]
}

df = pd.DataFrame(data)

# Define the threshold
threshold = 30

# Filter rows where 'Age' is greater than threshold
filtered_df = df[df['Age'] > threshold]

# Display the filtered DataFrame
print(filtered_df)
```

```python
11. # Perform matrix multiplication using NumPy?

import numpy as np

# Define two matrices (2D arrays)
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

# Method 1: Using np.dot()
result_dot = np.dot(A, B)

# Method 2: Using @ operator (Python 3.5+)
result_at = A @ B

print("Result using np.dot():\n", result_dot)
print("\nResult using @ operator:\n", result_at)

```

```python
12. # Use Pandas to load a CSV file and display its first 5 rows?


import pandas as pd

# Step 1: Create some sample data as a dictionary
data = {
    'Name': ['india', 'boil', 'oil', 'tin', 'Ev'],
    'Age': [25, 30, 35, 40, 28],
    'Score': [88, 92, 85, 95, 90]
}

# Step 2: Convert the dictionary to a DataFrame
df = pd.DataFrame(data)

# Step 3: Save the DataFrame to a CSV file
csv_filename = 'sample_data.csv'
df.to_csv(csv_filename, index=False)  # index=False to avoid saving the row numbers

print(f"CSV file '{csv_filename}' created.")

# Step 4: Load the CSV file back into a DataFrame
loaded_df = pd.read_csv(csv_filename)

# Step 5: Display the first 5 rows of the loaded DataFrame
print("First 5 rows of the loaded CSV:")
print(loaded_df.head())


```

```python
13. # Create a 3D scatter plot using Plotly.

import plotly.express as px
import pandas as pd

# Sample data
data = {
    'x': [1, 2, 3, 10, 5],
    'y': [10, 11, 12, 13, 14],
    'z': [5, 7, 5, 2, 1],
    'category': ['A', 'B', 'A', 'B', 'A']
}

df = pd.DataFrame(data)

# Create 3D scatter plot
fig = px.scatter_3d(df, x='x', y='y', z='z', color='category',
                    title='3D Scatter Plot Example')

# Show plot
fig.show()
```
