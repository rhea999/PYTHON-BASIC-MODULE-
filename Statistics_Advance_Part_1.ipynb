Q1: What is a random variable in probability theory?

A random variable is a variable that takes on numerical values determined by the outcome of a random experiment. It assigns a number to each possible outcome in a sample space.

Think of it as a function that maps outcomes to numbers.

For example, when you toss a coin, define a random variable X such that:

X = 1 if heads

X = 0 if tails

Even though the outcome is random, the random variable helps you quantify it and use mathematical tools to analyze probabilities.

Q2: What are the types of random variables?

There are two main types of random variables:

Discrete Random Variable:

Takes countable values (finite or countably infinite).

Examples: number of heads in 3 coin tosses (0, 1, 2, 3), number of students in a class.

Associated with probability mass functions (PMF).

Continuous Random Variable:

Takes infinite and uncountable values, usually over an interval.

Examples: height of students, time taken to complete a task.

Associated with probability density functions (PDF).

These types help determine how to compute and interpret probabilities depending on the nature of the outcomes.

Q3: What is the difference between discrete and continuous distributions?

* A discrete distribution deals with countable outcomes.
Example: Rolling a die (outcomes: 1, 2, 3, 4, 5, 6).
Each value has a specific probability assigned.

* A continuous distribution deals with uncountably infinite outcomes, often over an interval.
Example: Measuring height or time.
Individual values have zero probability, but intervals have probabilities (e.g., P(1.5 â‰¤ X â‰¤ 2.5)).

Key Difference:

* Discrete â†’ uses Probability Mass Function (PMF)

* Continuous â†’ uses Probability Density Function (PDF)

Q13: What is the Central Limit Theorem (CLT), and why is it critical in statistics?
The Central Limit Theorem (CLT) states that the sampling distribution of the sample mean approaches a normal distribution as the sample size increases, regardless of the populationâ€™s original distribution.

Why it's critical:

* Allows use of normal-based methods even when the population is not normal
* Justifies inferential statistics (e.g., confidence intervals, hypothesis testing)
* Makes it possible to predict population behavior from sample statistics

Q14: How does the Central Limit Theorem relate to the normal distribution?

The Central Limit Theorem (CLT) connects to the normal distribution by stating that:

As the sample size increases, the distribution of the sample mean becomes approximately normal, even if the population distribution is not normal.

In short:
CLT explains why normal distribution appears so often in statistics â€” because sample means tend to be normally distributed under many conditions.

Q17: What are point estimates and interval estimates in statistics?

* A point estimate gives a single best guess for a population parameter (like the sample mean estimating the population mean).

* An interval estimate provides a range of values within which the parameter is likely to lie, often with a confidence level.

Example:

* Point estimate: Mean height = 165 cm
* Interval estimate: Mean height is between 160â€“170 cm with 95% confidence

Q18: What is the significance of confidence intervals in statistical analysis?

A confidence interval (CI) gives a range of values within which the true population parameter is likely to fall, based on a sample.

Why it's significant:

* Reflects the uncertainty in sample estimates
* Provides a probabilistic range, not just a point
* Commonly used in reporting survey results, experimental outcomes, etc.

Example:
A 95% confidence interval means:
"If we repeated the sampling many times, 95% of the intervals would contain the true population mean."

```python
# Q2: Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Parameters
low, high = 1, 7  # simulating a 6-sided die

x = np.arange(low, high)
pmf = [1/(high - low) for _ in x]

plt.stem(x, pmf)
plt.title("PMF of Discrete Uniform Distribution (1 to 6)")
plt.xlabel("Outcome")
plt.ylabel("Probability")
plt.grid()
plt.show()
```

```python
# Q3: Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.
from scipy.stats import bernoulli

def bernoulli_pdf(p):
    x = [0, 1]
    pdf_values = bernoulli.pmf(x, p)
    return dict(zip(x, pdf_values))

# Example
print("Bernoulli PDF (p=0.6):", bernoulli_pdf(0.6))

```

```python
# Q15: Implement hypothesis testing using Z-statistics for a sample dataset
import numpy as np
from scipy.stats import norm

# Hypothesis: Î¼ = 50
sample = np.random.normal(52, 5, 30)
mu = 50
sigma = 5
n = len(sample)
sample_mean = np.mean(sample)

z = (sample_mean - mu) / (sigma / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z)))

print(f"Sample Mean = {sample_mean:.2f}")
print(f"Z-statistic = {z:.2f}")
print(f"P-value = {p_value:.4f}")

```

```python
# Q16: Create a confidence interval for a dataset using Python and interpret the result

import numpy as np
from scipy.stats import norm

data = np.random.normal(100, 15, 50)
mean = np.mean(data)
std = np.std(data, ddof=1)
n = len(data)

z = norm.ppf(0.975)  # 95% confidence
margin_error = z * (std / np.sqrt(n))

ci_lower = mean - margin_error
ci_upper = mean + margin_error

print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
# Q17: Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean

import numpy as np
from scipy.stats import norm

data = np.random.normal(loc=60, scale=10, size=100)
mean = np.mean(data)
std = np.std(data, ddof=1)
n = len(data)

z = norm.ppf(0.975)
ci = (mean - z * std/np.sqrt(n), mean + z * std/np.sqrt(n))

print(f"Sample Mean: {mean:.2f}")
print(f"95% Confidence Interval: {ci}")
```
