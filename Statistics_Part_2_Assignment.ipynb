### Q1. What is hypothesis testing in statistics?



_Answer:_
Hypothesis testing is a statistical method used to determine if there is enough evidence in a sample data to draw conclusions about a population. It involves formulating a null hypothesis (H0) and an alternative hypothesis (H1), collecting data, and using statistical tests to assess the likelihood of observing the data under the assumption that the null hypothesis is true.


### Q2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

_Answer:_
* Null Hypothesis (H₀): A statement that assumes no effect or no difference. It is the default assumption.

* Alternative Hypothesis (H₁ or Ha): A statement that contradicts H₀, suggesting there is an effect or difference.

* The hypothesis test determines whether the evidence is strong enough to reject H₀ in favor of H₁.


### Q3. What is the significance level in hypothesis testing, and why is it important?

_Answer:_

The significance level, usually denoted by α (alpha), is the threshold used to decide whether to reject the null hypothesis in a hypothesis test.

🔹 Common Values:
* α = 0.05 → 5% risk

* α = 0.01 → 1% risk

This means you are willing to accept a 5% or 1% chance of incorrectly rejecting a true null hypothesis (Type I error).

* It controls the probability of making a false positive.

* It sets the cut-off point for the p-value:

* If p ≤ α, reject the null hypothesis.

* If p > α, fail to reject the null hypothesis.

* It balances the risk of being too strict or too lenient in making decisions based on sample data.

🔹 Example:
If α = 0.05, you're saying:

"I am okay with a 5% chance of wrongly rejecting the null hypothesis."

So, if the p-value from your test is 0.03, you reject the null because 0.03 < 0.05.

### Q4. What does a P-value represent in hypothesis testing?

_Answer:_

The p-value is the probability of obtaining test results at least as extreme as the ones observed, assuming the null hypothesis is true. It measures the strength of evidence against the null hypothesis.

### Q5. How do you interpret the P-value in hypothesis testing?

_Answer:_

* If p ≤ α, reject H₀ (significant result).

* If p > α, fail to reject H₀ (not significant).

* A smaller p-value indicates stronger evidence against the null hypothesis.

### Q6. What are Type 1 and Type 2 errors in hypothesis testing?

_Answer:_

* Type I Error (False Positive): Rejecting a true null hypothesis.

* Type II Error (False Negative): Failing to reject a false null hypothesis.

### Q7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

_Answer:_

* One-tailed test: Tests for an effect in one direction (e.g., μ > μ₀).

* Two-tailed test: Tests for an effect in either direction (e.g., μ ≠ μ₀).

### Q8. What is the Z-test, and when is it used in hypothesis testing?

_Answer:_

A Z-test is used when:

* The population variance is known.

* The sample size is large (n > 30).
* It determines whether the sample mean differs significantly from the population mean.

### Q9. How do you calculate the Z-score, and what does it represent in hypothesis testing?

_Answer:_

Formula: Z = (X - μ) / σ
Where:

* X = data point

* μ = population mean

* σ = population standard deviation
The Z-score shows how many standard deviations a data point is from the mean.

_Answer:_

The T-distribution is used when:

* It has heavier tails than the normal distribution.

_Answer:_

* One-sample T-test

It's used when σ is unknown.

### Q14. What is a confidence interval, and how is it used to interpret statistical results?

_Answer:_

A confidence interval (CI) provides a range of values within which we expect the true population parameter to fall, with a specified level of confidence (e.g., 95%).

IT IS USED AS:-

1. Decision-making: If a CI for a difference between means includes 0, it suggests no significant difference.

2. Estimation: Instead of a single guess (point estimate), it gives a more reliable range.

3. Hypothesis testing: Can be used as an alternative to p-values to check statistical significance.



### Q15. What is the margin of error, and how does it affect the confidence interval?

_Answer:_

Margin of Error (MoE) is the range above and below the sample estimate.
Larger margin → Wider CI → Less precision.
MoE = Z * (σ/√n)

EFFECT ON CONFIDENCE INTERVAL:-


* More precision (smaller MOE) → narrower confidence interval.

* Less precision (larger MOE) → wider confidence interval.

### Q16. How is Bayes' Theorem used in statistics, and what is its significance?

_Answer:_

Bayes’ Theorem allows us to update probabilities based on new evidence:
P(A|B) = [P(B|A) * P(A)] / P(B)
It's fundamental in Bayesian inference, where we combine prior beliefs with new data.

Significance in Statistics:
1. Updates beliefs with data: It lets you adjust your expectations as new data comes in.

2. Used in Bayesian statistics: An alternative to frequentist approaches.

3. Helpful for decision-making under uncertainty.

### Q19. What is the F-distribution, and when is it used in hypothesis testing?

_Answer:_

The F-distribution is a right-skewed distribution used in statistics to compare variances between two or more groups.

🔹 Properties:
* Only positive values (starts at 0).

* Defined by two parameters: degrees of freedom for numerator (df₁) and denominator (df₂).

* Used mainly in ANOVA and F-tests.

🔹 When is it used?
* In ANOVA to test if group means are significantly different.

* In F-tests to compare variances of two populations.

🔹 Example:
* You have two different teaching methods and want to test if the variance in student scores differs — the F-distribution helps analyze that.



### Q20. What is an ANOVA test, and what are its assumptions?

_Answer:_

ANOVA (Analysis of Variance) is a statistical method used to determine whether there are significant differences between the means of 3 or more groups.

🔹 Purpose:
* Instead of doing multiple t-tests (which increases error), ANOVA performs one overall test.

🔹 Assumptions:
* Independence of observations.

* Normality: Data in each group should be normally distributed.

* Homogeneity of variances: All groups should have equal variances.

🔹 Example:
* You want to test if students from 3 different schools score differently on a math test. ANOVA can check if there's a statistically significant difference in the mean scores.

### Q21. What are the different types of ANOVA tests?

_Answer:_

There are several types of ANOVA depending on your data structure:

1. One-Way ANOVA
* Tests one independent variable (factor) with multiple groups.

* Example: Comparing test scores of students in 3 classes.

2. Two-Way ANOVA

* Tests the effect of two independent variables.

* Can also test for interaction effects.

* Example: Studying the effect of teaching method and gender on exam scores.

3. Repeated Measures ANOVA

* Used when the same subjects are tested multiple times.

* Example: Measuring blood pressure of patients before, during, and after treatment.

### Q48. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

```python
import numpy as np
from scipy.stats import norm

# Sample and known population mean
sample = np.array([55, 60, 58, 53, 59])
pop_mean = 50

sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)
n = len(sample)

z = (sample_mean - pop_mean) / (sample_std / np.sqrt(n))
p = 2 * (1 - norm.cdf(abs(z)))

print(f"Z-score: {z:.2f}")
print(f"P-value: {p:.4f}")

if p < 0.05:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")

```

### Q51. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Z-test setup
sample = np.array([48, 50, 52, 49, 51])
pop_mean = 50
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)
n = len(sample)

z = (sample_mean - pop_mean) / (sample_std / np.sqrt(n))
p = 2 * (1 - norm.cdf(abs(z)))

# Plotting the standard normal distribution
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

plt.plot(x, y)
plt.fill_between(x, y, where=(x <= -1.96) | (x >= 1.96), color='red', alpha=0.5, label='Rejection Region')
plt.axvline(z, color='blue', linestyle='--', label=f'Z = {z:.2f}')
plt.title("Two-Tailed Z-Test")
plt.xlabel("Z-value")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()

print(f"Z-score: {z:.2f}")
print(f"P-value: {p:.4f}")

```

### Q52. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_errors(mu0=0, mu1=1, sigma=1, alpha=0.05):
    x = np.linspace(-4, 4, 1000)
    null_dist = norm(mu0, sigma)
    alt_dist = norm(mu1, sigma)

    critical_value = norm.ppf(1 - alpha, mu0, sigma)

    plt.plot(x, null_dist.pdf(x), label='Null Hypothesis (H₀)')
    plt.plot(x, alt_dist.pdf(x), label='Alternative Hypothesis (H₁)')
    plt.axvline(critical_value, color='red', linestyle='--', label='Critical Value')

    # Type I error region
    plt.fill_between(x, 0, null_dist.pdf(x), where=(x > critical_value), color='orange', alpha=0.5, label='Type I Error')
    # Type II error region
    plt.fill_between(x, 0, alt_dist.pdf(x), where=(x < critical_value), color='blue', alpha=0.3, label='Type II Error')

    plt.title("Type I and Type II Errors")
    plt.xlabel("Test Statistic")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

visualize_errors()

```

### Q53. Write a Python program to perform an independent T-test and interpret the results.

```python
from scipy.stats import ttest_ind

group1 = np.random.normal(loc=50, scale=5, size=30)
group2 = np.random.normal(loc=52, scale=5, size=30)

t_stat, p_val = ttest_ind(group1, group2)

print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_val:.4f}")

if p_val < 0.05:
    print("Reject the null hypothesis: Groups are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference.")

```

### Q55. Simulate data and perform both Z-test and T-test, then compare the results using Python.

```python
import numpy as np
from scipy.stats import norm, ttest_1samp

# Sample data
sample = np.random.normal(loc=51, scale=5, size=30)
pop_mean = 50

# Z-test
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)
z = (sample_mean - pop_mean) / (sample_std / np.sqrt(len(sample)))
z_p_value = 2 * (1 - norm.cdf(abs(z)))

# T-test
t_stat, t_p_value = ttest_1samp(sample, popmean=pop_mean)

# Print comparison
print("Z-test:")
print(f"  Z-score: {z:.2f}, P-value: {z_p_value:.4f}")

print("\nT-test:")
print(f"  T-statistic: {t_stat:.2f}, P-value: {t_p_value:.4f}")

```

### Q56. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    mean = np.mean(data)
    sem = np.std(data, ddof=1) / np.sqrt(len(data))
    ci = t.interval(confidence, df=len(data)-1, loc=mean, scale=sem)
    return mean, ci

data = np.random.normal(100, 10, 30)
mean, ci = confidence_interval(data)

print(f"Sample Mean: {mean:.2f}")
print(f"{int(100*0.95)}% Confidence Interval: {ci}")

```

### Q57. Write a Python program to calculate the margin of error for a given confidence level using sample data.

```python
from scipy.stats import norm

def margin_of_error(data, confidence=0.95):
    z = norm.ppf((1 + confidence) / 2)
    sem = np.std(data, ddof=1) / np.sqrt(len(data))
    return z * sem

data = np.random.normal(50, 5, 40)
moe = margin_of_error(data)
print(f"Margin of Error (95%): {moe:.2f}")

```

### Q58. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

```python
def bayes_theorem(prior_A, prob_B_given_A, prob_B):
    posterior = (prob_B_given_A * prior_A) / prob_B
    return posterior

# Example: Medical test
# P(Disease) = 0.01, P(Positive | Disease) = 0.99, P(Positive) = 0.05
prior = 0.01
likelihood = 0.99
evidence = 0.05

posterior = bayes_theorem(prior, likelihood, evidence)
print(f"Posterior Probability (Having Disease | Positive Test): {posterior:.4f}")

```

### Q60. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.

```python
import numpy as np

# Observed data
observed = np.array([[30, 10],
                     [20, 40]])

# Row and column sums
row_totals = observed.sum(axis=1)
col_totals = observed.sum(axis=0)
total = observed.sum()

# Expected frequency matrix
expected = np.outer(row_totals, col_totals) / total
print("Expected Frequencies:")
print(expected)

```

### Q61. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.

```python
from scipy.stats import chisquare

# Observed frequencies (e.g., how many times each side appeared)
observed = [8, 12, 10, 11, 9, 10]

# Expected frequencies (assuming fair die → each side appears 10 times)
expected = [10, 10, 10, 10, 10, 10]

# Perform Chi-square goodness-of-fit test
chi2_stat, p_val = chisquare(f_obs=observed, f_exp=expected)

print(f"Chi-square Statistic: {chi2_stat:.2f}")
print(f"P-value: {p_val:.4f}")

# Interpretation
if p_val < 0.05:
    print("Result: Reject the null hypothesis. The die may not be fair.")
else:
    print("Result: Fail to reject the null hypothesis. The die appears fair.")


```
